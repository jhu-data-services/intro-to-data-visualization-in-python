## Introduction to Matplotlib
`matplotlib` is a Python library for producing publication-quality figures.

There are two interfaces to `matplotlib`;
1. ``matplotlib.pyplot`` and 
2. An *Object Oriented* interface.

The `pyplot` interface provides a simpler (but less flexible) means of interfacing with `matplotlib`. 

We will explore the `pyplot` interface first, identify some of its limitations, and then explore the addtional flexibility and functionality provided by the *Object Oriented* interface.

### Importing `matplotlib.pyplot`

```python
import matplotlib.pyplot as plt
# Use rcParams to change properties for your figures GLOBALLY
# This means that every figure will be 10 x 5. 
plt.rcParams['figure.figsize'] = (10, 5)

# To reset to the default size use
# plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]

# Alternatively you can set individual figure dimensions using figsize.
# The example below sets the figure to an 800 x 600 dimension (fig size * dpi)
# figure(figsize=(8, 6), dpi=100)
```

`matplotlib.pyplot` is a collection of functions, known as a submodule, that make `matplotlib` work like the `MATLAB` `plot` function. We import `matplotlib.pyplot` to the alias `plt` as a matter of common convention.

Each pyplot function makes some change to a figure, for example: 
- creates a figure
- creates a plotting area in a figure 
- plots some lines or points in a plotting area
- decorates the plot labels

### Anatomy of a Matplotlib Figure

When we called the `plt.plot` function, and later the `plt.scatter` function, a couple of things were happening in the background: 

- matplotlib created a **figure** instance; the figure is a container which holds everything you see. 
- matplotlib created an **axes** instance; the axes are the part of the figure that holds the data. This is the 'canvas' on which we will paint with our data.

These two pieces make up our matplotlib figures:

<img src='https://matplotlib.org/stable/_images/anatomy.png' alt='matplotlib fig anatomy' width='600'>

Image Source: https://matplotlib.org/stable/_images/anatomy.png

The **figure** object (`fig`) is a container which holds everything you see. This is the top level container for all other elements that make up the graphic.

The blue elements are called **artists**. The Figure object can be thought of as a canvas, upon which different artists act to create the final graphic image. 

Each plot we see in a figure is an **axes** object. We can have one **figure** (container) but multiple **axes** (plots).

```python
fig = plt.figure()
plt.show()
```

Notice that when we run this plot, nothing is actually produced. This is because we haven't put anything into our container yet! 

The **axes** (`ax`) is the part of the figure that holds the data. This is the 'canvas' on which we will paint with our data.

```python
fig, ax = plt.subplots()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
plt.show()
```

### Matplotlib Object Oriented Interface

Now, let's re-build our plot with the Object Oriented API.

```python
fig, ax = plt.subplots()
ax.scatter(x='body_mass_g', 
           y='bill_length_mm',
           alpha=0.5,
           s=60,
           c=species_colors,
           data=penguins)
ax.set_title('Bill Length and Bill Depth', fontsize=15)
ax.set_xlabel('Body Mass (g)', fontsize=14)
ax.set_ylabel('Bill Length (mm)', fontsize=14)
ax.tick_params(axis='both', which='minor', labelsize=12)

ax.spines[['top', 'right']].set_visible(False)
plt.show()
```

We used `plt.subplots` which will allow us to return multiple axes for a single figure. By default only one axes is displayed, although we can specify more:

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
plt.show()
```

This just seems like a more confusing way of creating the same plot. Why would we want to do this?

We can actually plot multiple axes. We are already comparing **Body Mass** on the x-axis and **Bill Length** on the y-axis.
Lets add another y-axis with **Bill Depth** as well!

```python
fig, ax = plt.subplots(nrows=1,ncols=2)

# When we have more than one axis, we reference it with an index
# fig, ax = plt.subplots(nrows=1,ncols=2)
# ax[0] to reference the first axis, and ax[1] the second

# When we have multiple rows and columns we need both a row and column index
# fig, ax = plt.subplots(nrows=2, ncols=2)
# ax[0,0] to access the first column and first row axis
ax[0].scatter(x='body_mass_g', 
           y='bill_length_mm',
           alpha=0.5,
           s=60,
           c=species_colors,
           data=penguins)
ax[0].set_title('Bill Length and Bill Depth', fontsize=15)
ax[0].set_xlabel('Body Mass (g)', fontsize=14)
ax[0].set_ylabel('Bill Length (mm)', fontsize=14)
ax[0].tick_params(axis='both', which='minor', labelsize=12)
ax[0].spines[['top', 'right']].set_visible(False)

ax[1].scatter(x='bill_length_mm', 
              y='flipper_length_mm',
              alpha=.5,
              data=penguins)
ax[1].set_title('Flipper Length and Bill Length', fontsize=15)
ax[1].set_xlabel('Bill Length (mm)',  fontsize=14)
ax[1].set_ylabel('Flipper Length (mm)', fontsize=14)
ax[1].spines[['top', 'right']].set_visible(False)
plt.show()
```

#### Add title over all subfigures

```python
fig, ax = plt.subplots(nrows=1,ncols=2)

ax[0].scatter(x='body_mass_g', 
           y='bill_length_mm',
           alpha=0.5,
           s=60,
           c=species_colors,
           data=penguins)
ax[0].set_title('Bill Length and Bill Depth', fontsize=15)
ax[0].set_xlabel('Body Mass (g)', fontsize=14)
ax[0].set_ylabel('Bill Length (mm)', fontsize=14)
ax[0].tick_params(axis='both', which='minor', labelsize=12)
ax[0].spines[['top', 'right']].set_visible(False)

ax[1].scatter(x='bill_length_mm', 
              y='flipper_length_mm',
              alpha=0.5,
              s=60,
              c=species_colors,
              data=penguins)
ax[1].set_title('Flipper Length and Bill Length', fontsize=15)
ax[1].set_xlabel('Bill Length (mm)',  fontsize=14)
ax[1].set_ylabel('Flipper Length (mm)', fontsize=14)
ax[1].spines[['top', 'right']].set_visible(False)

# Add title to the whole figure
fig.suptitle('Palmer Penguin Characteristics', fontsize=16)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax[0].scatter(x='body_mass_g', 
           y='bill_length_mm',
           alpha=0.5,
           s=60,
           c=species_colors,
           data=penguins)
ax[0].set_title('Bill Length and Bill Depth', fontsize=15)
ax[0].set_xlabel('Body Mass (g)', fontsize=14)
ax[0].set_ylabel('Bill Length (mm)', fontsize=14)
ax[0].tick_params(axis='both', which='minor', labelsize=12)
ax[0].spines[['top', 'right']].set_visible(False)

ax[1].scatter(x='bill_length_mm', 
              y='flipper_length_mm',
              alpha=0.5,
              s=60,
              c=species_colors,
              data=penguins)
ax[1].set_title('Flipper Length and Bill Length', fontsize=15)
ax[1].set_xlabel('Bill Length (mm)',  fontsize=14)
ax[1].set_ylabel('Flipper Length (mm)', fontsize=14)
ax[1].spines[['top', 'right']].set_visible(False)
fig.suptitle('Palmer Penguin Characteristics', fontsize=16)


plt.savefig("palmer_penguin_characteristics.pdf", dpi=100)
```

### Seaborn Scatterplot

Recreate a scatter plot that compares `body mass` on the x-axis to `bill length` on the y-axis using `seaborn`.

We will want to:
* Add axis labels
* Scale the font size to make it more readable
* Remove the extraneous spines

```python
import seaborn as sns
sns.scatterplot(x = 'body_mass_g',
                y = 'bill_length_mm',
                data=penguins)
plt.show()
```

### Seaborn Plot Themes
By default, seaborn plots will be made with the current values of `matplotlib`. If you have already configured the style of your matplotlib (as we have above) that theme will be continue to be used. We can change the seaborn theme with `sns.set_theme(style = 'style type')`. Calling this function with no arguments will activate seaborn’s “default” theme:

```python
sns.set_theme()
plt.figure(figsize = (8, 6))
sns.scatterplot(x = 'body_mass_g',
                y = 'bill_length_mm',
                data=penguins)
plt.show()
```

Seaborn has five themes built in: `darkgrid` , `whitegrid` , `dark` , `white` , and `ticks`. Let's use `ticks` which will apply axis ticks with no background color or grid.

```python
sns.set_theme(style = 'ticks')
sns.scatterplot(x = 'body_mass_g',
                y = 'bill_length_mm',
                data=penguins)
fig_sav = plt.gcf()
```

Add axis labels:

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='body_mass_g',
               y='bill_length_mm',
               data=penguins)
plt.xlabel('Body Mass (g)')
plt.ylabel('Bill Length (mm)')
plt.show()
```

Scale the font size

```python
sns.set_context('paper', font_scale=1.5)
sns.scatterplot(x='body_mass_g',
               y='bill_length_mm',
               data=penguins)

plt.xlabel('Body Mass (g)')
plt.ylabel('Bill Length (mm)')
plt.show()
```

Remove the spines

```python
sns.scatterplot(x='body_mass_g',
               y='bill_length_mm',
               s=60,
               alpha=0.5,
               data=penguins)
sns.despine()
plt.xlabel('Body Mass (g)')
plt.ylabel('Bill Length (mm)')
plt.show()
```

### Color by species
Now that we have a similar plot to that from our `Matplotlib` section, lets also add color, by species:

```python
sns.scatterplot(x='body_mass_g',
               y='bill_length_mm',
               s=60,
               alpha=0.5,
               hue='species',
               data=penguins)
sns.despine()
plt.xlabel('Body Mass (g)')
plt.ylabel('Bill Length (mm)')
plt.show()
```

### Seaborn Regression Plot

```python
sns.lmplot(x = 'body_mass_g',
           y = 'bill_length_mm',
           hue = 'species',
           height = 7,
           data=penguins)
sns.despine()
plt.xlabel('Body Mass (g)')
plt.ylabel('Bill Length (mm)')
plt.show()
```

### Seaborn Joint Plot

```python
sns.jointplot(x = 'body_mass_g',
           y = 'bill_length_mm',
           hue = 'species',
           s=60,
           alpha=0.5,
           height = 7,
           data = penguins,
           kind = 'scatter')
plt.xlabel('Body Mass (g)')
plt.ylabel('Bill Length (mm)')
plt.show()
```

### Seaborn Pair Plot

```python
sns.pairplot(penguins, hue='species')
plt.show()
```

### Seaborn Histogram

```python
sns.histplot(x='bill_length_mm',
             data=penguins)
plt.xlabel('Bill Length (mm)')
plt.show()
```

Add a Kernel Density Estimate (KDE)

```python
sns.histplot(x='bill_length_mm',
             data=penguins,
             kde=True)
plt.xlabel('Bill Length (mm)')
plt.show()
```

Vary the number of bins

```python
sns.histplot(x='bill_length_mm',
             data=penguins,
             kde=True,
             bins=25)
plt.xlabel('Bill Length (mm)')
plt.show()
```
